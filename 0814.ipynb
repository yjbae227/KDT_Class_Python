```python
# Python이 제공하는 Built-in data type
# 1. Numeric : 모든 숫자 (클래스로 숫자의 종류 구분 -> 정수, 실수, 복소수)
```

```python
# Numeric Type (숫자형) --> Data types
# int (정수)            --> class
# float (실수)          --> class
# complex (복소수)      --> class
a = 100
print(a)
print(type(a))          # <class 'int'>
b = 3.14
print(type(b))          # <class 'float'>
# Python의 모든 값은 객체 -> 객체는 필수적으로 class로부터 생성됨(파생)
# 간단한 연산
div = 3 / 4
print(div)              # 결과는 -> 0.75
# 프로그래밍에서 연산은 동타입만 수행 됨
#                                        -> 결과도 같은 동타입으로 발생
#                                        ==> 기본 연산 수행 시 모든 숫자가 실수로 자동변환 후 수행
result = 3 ** 3        # ** : 지수표현 (제곱변)
print(result)
result = 10 % 3        # % : 나머지 연산
print(result)
result = 10 //3        # // : 몫 연산
print (result)
```

```python
# Python의 *** List *** : []
# 가장 기본적, 대표적, 중요한 자료구조이면서 데이터 타입
# list 는 class 명 / ','로 객체 구분 가능
# 임의의 객체를 순차적으로 저장하는 집합형 자료구조 (like 배열)
#                                                  -> Python에는 array가 없음
# literal은 []를 이용                      -> literal : 약속된 기호
a = []                              # empty List 생성
a = [1, 2, 3, 4, 5]                 # 5개의 int객체가 순차적으로 저장되어있는 lIst 생성
a = [1, 2.0, 3.14, 'Hello',[3, 5]]      # int, float, float, 문자열, list ==> 중첩개념

a = [1, 2, ['Show', 'me', 'the', 'money!'], 3.14]
print(a)
# list를 사용하기 위해서 indexing과 slicing에 대한 이해 필요
# indexing은 순서를 이용해서 값을 알아내거나 값을 변경하는 방법을 의미
# indexing을 할 때 []를 이용
print(a[1])                        # 2
print(a[2][0])                     # Show
print(a[-1])                       # Python은 minus index를 제공 -> 오른쪽기준 indexing
# *** Slicing ***  : [n:m] -> n은 포함하고 m은 불포함한 slicing
# slicing의 결과는 원본과 같은 데이터 타입으로 생성
print(a[0:2])                     # [1, 2]
print(a[1:])                      # 앞의 숫자가 생략되면 처음부터, 뒤 숫자가 생략되면 끝까지
print(a[:])                       # 처음부터 끝까지
```

```python
# Sequence Type의 ***list***
# list의 연산
a = [1, 2, 3]
b = [4, 5, 6]
print (a + b)                           # [1, 2, 3, 4, 5]
# list의 + 연산은 연결연산
print (a * 3)                           # [1, 2, 3, 1, 2, 3, 1, 2, 3]
# 파이썬에서 나오는 모든 값은 객체
# 객체는 반드시 class부터 파생
# 객체는 도대체 어떤 건가요? => 값과 함수(기능)가 같이 결함되어 있는 자료구조
a.append(4)
print(a)                               # [1, 2, 3, 4]

mylist = [4, 6, 8, 3, 10, 9, 1]
mylist.sort()                        # 오름차순 정렬
mylist.reverse()                       # 순서를 역으로 돌림(원점대칭)
print(mylist)

# 객체가 가지는 기능(함수 : method)는 두 가지 형태가 존재
# 1. 객체가 가지는 데이터를 직접 변화시키는 함수 (ex. sort)
# 2. 객체가 가지는 데이터는 변화시키지 않으며 결과를 도출하는 함수 --> return이라서 객체 생성 후 받아와야 함  (ex. )
```

```python
#
#
#
#
#
# Sequence *** tuple *** : list와 유사한 순서가 있는 집합형 자료구조
#                        () 사용
a = ()              #  빈 tuple 생성
print(type(a))      # <class 'tuple'>
# 요소가 1개인 tuple
a = (1,)
print(type(a))      # <class 'tuple'>
a = (1, 2, 3, 4)    # 일반적인 형태의 tuple
# tuple은 소괄호를 생략 할 수 있음 -> 대부분 생략해서 사용
a = 1, 2, 3, 4
# 파이썬은 아래와 코드가 많이 보임 (다중생성,설정)
a, b, c = 10, 20, 30
# list / tuple의 차이 => tuple : read only
a = [1, 2, 3, 4]
a[0] = 100
print(a)           # [100, 2, 3, 4]
a = 1, 2, 3, 4
# a[0] = 100         # 값 수정 불가 -> 'tuple' object does not support item assignment
```

```python
# list와 tuple은 class
a = []                      # 비어있는 list
a = list()                  # 생성자 함수
b = ()                      # 비어있는 tuple
b = tuple()                 # 생성자 함수

# 일반적으로 다음과 같이 사용한다.
a = [1, 2, 3, 4]
b = tuple(a)
print(b)                    # (1, 2, 3, 4)
```

```python
# Sequence Type *** range ***
#
#
# range(범위)
# list와 tuple은 실제 데이터에 메모리를 가지고 있음

a = [1, 2, 3, 4, 5, 6, 7]
range1 = range(10, 100)          # 값의 범위를 가지고 있는 객체
print(range1)                   # range(1, 100)
# 그러면 indexing은 되나요? -> 가능
print(range1[2])                # 12
# 그러면 slicing은 되나요? -> 가능
print(range1[2:10])
```

```python
# Python이 제공하는 Built-in data type
# 1. Numeric
# 2. Sequence
# 3. Text Sqcuenct Type (문자열)
#   - str




# Python의 문자열
# 문자와 문자열의 구분이 없음
# 일반적으로 다른 언어에서 문자는 1글자 지칭 => ''
# 문자열은 여러글자를 지칭함 => ""
# Python은 모두 문자열이에요, 따라서 '', ""의 차이가 존재하지 않음
#           -> 관용적으로 ''를 더 많이 사용함.
a = "안녕하세요"
b = '하이하이!!!'
c = '''이것은
소리없는
아우성'''
print(c)

# 문자열은 list안에 각각의 문자들이 들어가 있는 형태
first = '이것은 '
second = '소리없는 '
third = '아우성'
print(first + second + third)
```

```python
# Python이 제공하는 Built-in data type
# 1. Numeric
# 2. Sequence
# 3. Text Sqcuenct Type (문자열)
# 4. Mapping type
#       - dict







# dict(class)를 이용해서 만든 객체를 python에서는
# dictionary라고 불러요
# key와 valuce의 쌍으로 데이터를 표현
# literal은 {}를 이용
a = {}              # empty dictinary
print(type(a))      # <class 'dict'>
a = {'name' : '홍길동'}
# 그러다 보니 value에 해당하는 값은 python에서 나오는 값으 ㄹ
# 사용할 수 있어요!
# 하지만 key는 변경이 안되는 값을 이용해야만 해요
# key값으로는 list가 올 수 없어요
# 만약 데이터를 어려개 저장하려면
student = {
    'name': '홍길동',
    'dept': '컴퓨터',
    'socre' : 3.6
}
# indexcing이 가능함, 단 key 값을 이용해서 사용해야 함
print(student['dept'])

# 이런 경우는 어떻게 되요?   => 코드상 문제는 없지만
# 당연히 사용하지 않아요
student = {
    'name' : '홍길동',
    # 'name' : '신사임당',
    'dept' : '컴퓨터',
    'score' : 3.6,
    ('address',) : '서울'
}

student[('address',)] = '부산'
print(student[('address', )])
```

```python
# dictionary의 대표적 함수
student = {
    'name' : '홍길동',
    'dept' : '컴퓨터',
    'score' : 3.6
}

# 1. keys()
print(student.keys())           # dict_keys(['name', 'dept', 'score'])
print(type(student.keys()))     # <class 'dict_keys'>
# 2. values()
print(student.values())
```
